```python
!pip install playsound

import cv2
import numpy as np
from tensorflow.keras.models import load_model
from collections import deque
from playsound import playsound
import threading  # To play sound without freezing video

# Load the trained model
model = load_model("drowsiness_detection_model.keras")
print("✅ Model loaded successfully! Ready for real-time detection.")

# Define the sequence length
seq_length = 10
frame_queue = deque(maxlen=seq_length)

# Start webcam
cap = cv2.VideoCapture(0)

# Function to play alarm in a separate thread
def play_alarm():
    playsound("alarm.mp3")  # Make sure alarm.mp3 is in the same directory

try:
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Resize and normalize frame
        processed_frame = cv2.resize(frame, (224, 224)) / 255.0
        frame_queue.append(processed_frame)

        if len(frame_queue) == seq_length:
            input_sequence = np.array(frame_queue).reshape(1, seq_length, 224, 224, 3)

            # Predict drowsiness
            prediction = model.predict(input_sequence)
            label = "Non Drowsy" if np.argmax(prediction) == 1 else "Drowsy"

            # Play alarm if drowsy (in a separate thread to avoid freezing)
            if label == "Drowsy":
                threading.Thread(target=play_alarm, daemon=True).start()

            # Display the result on video frame
            cv2.putText(frame, label, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        # Show the video
        cv2.imshow("Driver Drowsiness Detection", frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

except KeyboardInterrupt:
    print("\n🛑 Keyboard Interrupt detected. Closing application...")

finally:
    cap.release()
    cv2.destroyAllWindows()
    print("✅ Resources released. Program exited safely.")
```
